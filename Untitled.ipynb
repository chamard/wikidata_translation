```python
revisions_df = pd.read_csv(f'500_revisions.csv')
```

```python
# We will only work on the label creation for the momment
label_creation_df = revisions_df.loc[((~revisions_df.label.isna())
                                      & (revisions_df.action == 'add')),
                                     ["property","language","timestamp"]]
```

```python
languages = label_creation_df.language.unique()
int_to_language = {i:el for i,el in enumerate(languages)}
language_to_int = {el:i for i,el in int_to_language.items()}
```

```python
label_creation_df['language'] = label_creation_df.language.astype(str)
```

```python
label_creation_df.sort_values(["property","timestamp"],inplace=True)
```

```python
last_prop,i,position = None,1,[]

for index, row in label_creation_df.copy().iterrows():
    if(row['property'] != last_prop):
        i, last_prop = 1, row['property']
    position.append(i)
    i += 1
```

```python
label_creation_df['position'] = position
```

```python
revisions_history_df= pd.pivot_table(data=label_creation_df,
                                     index="property",
                                     columns="position",
                                     values='language',
                                     aggfunc=lambda x: ' '.join(x))
```

```python
revisions_history_df.fillna('',inplace=True)
```

```python
# Create the history
history = []
for _,row in revisions_history_df.iterrows():
    history.append([language_to_int[el] for el in row.values if el])
    
```

```python
from orangecontrib.associate.fpgrowth import *  
itemsets = frequent_itemsets(history)
itemsets = dict(frequent_itemsets(history, .5))
#print(list(itemsets))

print("")
rules = association_rules(itemsets, 0.9)
test = list(rules)
print(test)
```
