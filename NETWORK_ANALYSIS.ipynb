## Language code labels

```python
sparql = SPARQLWrapper("https://query.wikidata.org/sparql")
sparql.setQuery("""SELECT ?c ?label
{
  ?item wdt:P424 ?c;
        rdfs:label ?label FILTER (lang(?label) = "en").
  MINUS{?item wdt:P31/wdt:P279* wd:Q14827288} #exclude Wikimedia projects
  MINUS{?item wdt:P31/wdt:P279* wd:Q17442446} #exclude Wikimedia internal stuff
}
ORDER by ?c""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
```

## Translations dataframe

```python
last_prop, last_language, translations = None, None, []

for index, row in label_creations_df.iterrows():
    language = row['language']
    if(row['property'] != last_prop):
        last_prop = row['property']
    else:
        edge = (last_language,language)
        translations.append(edge)
    
    last_language = language
```

```python
translations_df = pd.DataFrame(translations)
translations_df.rename(columns={0: "Source", 1: "Destination"},inplace=True)
translations_df['Count'] = 1
translations_count = translations_df.groupby(['Source','Destination']).Count.count().reset_index()
```
